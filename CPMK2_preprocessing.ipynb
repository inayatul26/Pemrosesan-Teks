```python
#import library
from google_play_scraper import Sort, reviews
from google_play_scraper import app
import pandas as pd
import numpy as np

```

```python
#mulai scraping
result, continuation_token = reviews(
    'id.co.bankbkemobile.digitalbank',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=1000,
    filter_score_with=None
)
```

```python
#membuat DataFrame dari hasil scraping
df = pd.DataFrame(np.array(result), columns=['review'])
df = df.join(pd.DataFrame(df.pop('review').tolist()))
df
```

```python
df['replyContent'].fillna('Tidak ada balasan', inplace=True)
df['repliedAt'].fillna('Tidak ada tanggal balasan', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['content_casefold'] = df['content'].astype(str).str.lower()
```

```python
!pip install emoji
import re
import emoji
def clean_text(text):
    # Hapus URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # Hapus emoji secara otomatis
    text = emoji.replace_emoji(text, replace='')  # hapus semua emoji

    # Hapus angka dan simbol
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['review'] = df['content_casefold'].apply(clean_text)
```

```python
import string
import nltk
from nltk.corpus import stopwords

# === 1. Unduh stopwords NLTK (jalankan sekali saja) ===
nltk.download('stopwords')
# === 2. Ambil stopwords Bahasa Indonesia ===
stop_words = set(stopwords.words('indonesian'))
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

stop_words = set(stopwords.words('indonesian'))
extra_stopwords = {'deh', 'dong', 'nih', 'sih', 'loh', 'ya', 'kan', 'kok', 'lah'}
all_stopwords = stop_words.union(extra_stopwords)

def remove_stopwords_all(text):
    words = text.split()
    filtered = [w for w in words if w not in all_stopwords]
    return ' '.join(filtered)

df['review'] = df['review'].apply(remove_stopwords_all)
```

```python
# Kamus normalisasi (kata typo, singkatan, gaul, dll)
normalisasi_dict = {
    'gk': 'tidak', 'ga': 'tidak', 'nggak': 'tidak', 'gak': 'tidak',
    'bgt': 'banget', 'bener': 'benar', 'trnyata': 'ternyata',
    'tdk': 'tidak', 'yg': 'yang', 'dlm': 'dalam', 'dr': 'dari',
    'sm': 'sama', 'aja': 'saja', 'udh': 'sudah', 'udah': 'sudah',
    'dgn': 'dengan', 'bisaaa': 'bisa', 'mantappp': 'mantap',
    'bgs': 'bagus', 'ok': 'oke', 'okee': 'oke', 'okey': 'oke',
    'trs': 'terus', 'krn': 'karena', 'blm': 'belum', 'pls': 'tolong',
    'smoga': 'semoga', 'makasih': 'terima kasih', 'makasi': 'terima kasih',
    'makasii': 'terima kasih', 'makasihh': 'terima kasih', 'thx': 'terima kasih',
    'tq': 'terima kasih', 'mantapp': 'mantap', 'mantul': 'mantap betul'
}

# Fungsi normalisasi lanjutan
def normalize_text(text):
    # Hapus huruf berulang (contoh: bagusss → bagus)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Normalisasi kata slang/singkatan
    words = text.split()
    normalized_words = [normalisasi_dict.get(w, w) for w in words]

    # Hapus kata yang panjangnya kurang dari 3 huruf
    filtered = [w for w in normalized_words if len(w) > 3]

    return ' '.join(filtered)

# Terapkan ke hasil clean_text kamu
df['review'] = df['review'].apply(normalize_text)
```
